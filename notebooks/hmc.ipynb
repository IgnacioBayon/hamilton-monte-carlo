```python
def f(x, a=1, b=100):
    """
    Rosenbrock function (Banana function)
    """
    return (a - x[0])**2 + b * (x[1] - x[0]**2)**2
```

```python
# plot contour of f
x_space, y_space = np.meshgrid(np.linspace(-2, 2, 1000), np.linspace(-2, 2, 1000))

Z_small = (1 - x_space)**2 + 100 * (y_space - x_space**2)**2

# plot contour
plt.contourf(x_space, y_space, Z_small, levels=np.logspace(-2, 3, 50), cmap='viridis')
plt.title('Rosenbrock function contour')
plt.xlabel('x1')
plt.ylabel('x2')
plt.colorbar()
plt.show()
```

```python
def grad_f(x, a=1, b=100):
    """
    Gradient of the Rosenbrock function
    """
    dfdx = -2 * (a - x[0]) - 4 * b * x[0] * (x[1] - x[0]**2)
    dfdy = 2 * b * (x[1] - x[0]**2)
    return dfdx, dfdy
```

```python
num_samples = 1000
L = 50
step_size = 0.01

samples = hamiltonian_sample(f, grad_f, num_samples, L, step_size)
```
